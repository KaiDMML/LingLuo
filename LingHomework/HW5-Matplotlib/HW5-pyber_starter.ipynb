```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
City = pd.read_csv(city_data_to_load)
Ride = pd.read_csv(ride_data_to_load)
# Combine the data into a single dataset
Combo = pd.merge(City, Ride, on="city")
# Display the data table for preview
Combo.head(10)
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
urban = Combo.loc[(Combo["type"] == "Urban")]
suburban = Combo.loc[(Combo["type"] == "Suburban")]
rural = Combo.loc[(Combo["type"] == "Rural")]

x1_axis=urban.groupby("city").count()["ride_id"]
y1_axis=urban.groupby("city").mean()["fare"]
x2_axis=suburban.groupby("city").count()["ride_id"]
y2_axis=suburban.groupby("city").mean()["fare"]
x3_axis=rural.groupby("city").count()["ride_id"]
y3_axis=rural.groupby("city").mean()["fare"]


# Build the scatter plots for each city types
plt.scatter(x1_axis,y1_axis,marker ='o', facecolors ="red", edgecolors='black',alpha = 0.7, label="Urban")
plt.scatter(x2_axis,y2_axis,marker ='o', facecolors ="lightblue", edgecolors='black',alpha = 0.7, label="Suburban")
plt.scatter(x3_axis,y3_axis,marker ='o', facecolors ="yellow", edgecolors='black',alpha = 0.7, label="Rural") 

# Incorporate the other graph properties
plt.xlabel("Total numbers of rides(per city)")
plt.ylabel("Average fare($)")
plt.title("Pyber Ride Sharing Data (2016)")

# Create a legend
plt.legend(loc='upper right')
# Incorporate a text label regarding circle size

# Save Figure
plt.savefig("../Images/Pyber Ride Sharing.png")
plt.show()
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
total_fare = Combo.groupby(['type'])['fare'].sum()
# Build Pie Chart
labels = ["Rural","Suburban","Urban" ]
# Save Figure
colors = ["red","blue","yellow"]
explode = (0, 0, 0.1)
plt.title("Percentage of Total Fares By City Types")
plt.pie(total_fare, explode=explode, labels=labels, colors=colors, shadow=True, startangle=160)
plt.axis("equal")
plt.savefig("../Images/Percnetage of Total Fares By City Types.png")
plt.show()
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
total_rides = Combo.groupby(['type'])['ride_id'].count()
# Build Pie Chart
labels = ["Rural","Suburban","Urban" ]
# Save Figure
colors = ["red","blue","yellow"]
explode = (0, 0, 0.1)
plt.title("Percentage of Total rides By City Types")
plt.pie(total_rides, explode=explode, labels=labels, colors=colors, shadow=True, startangle=160)
plt.axis("equal")
plt.savefig("../Images/Percnetage of Total rides By City Types.png")
plt.show()
```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
total_drivers = Combo.groupby(['type'])['driver_count'].sum()
labels = ["Rural","Suburban","Urban" ]
# Save Figure
colors = ["red","blue","yellow"]
explode = (0, 0, 0.1)
plt.title("Percentage of Total drivers By City Types")
plt.pie(total_drivers, explode=explode, labels=labels, colors=colors, shadow=True, startangle=160)
plt.axis("equal")
plt.savefig("../Images/Percnetage of Total rides By City Types.png")
plt.show()
```
